```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("US_Accidents_March23/US_Accidents_March23.csv")
```

```python
print("Shape:", df.shape)
print(df.head())
```

```python
df = df.dropna(subset=["Start_Lat", "Start_Lng"])
```

```python
df["Start_Time"] = pd.to_datetime(df["Start_Time"],errors='coerce')
df["Hour"] = df["Start_Time"].dt.hour
```

```python
# Top 10 Weather Conditions
plt.figure(figsize=(10,5))
sns.countplot(y="Weather_Condition", data=df,
              order=df["Weather_Condition"].value_counts().iloc[:10].index)
plt.title("Top 10 Weather Conditions for Accidents")
plt.xlabel("Number of Accidents")
plt.ylabel("Weather Condition")
plt.show()
```

```python
# Accidents by Hour
plt.figure(figsize=(8,5))
sns.countplot(x="Hour", data=df, color="orange")
plt.title("Accidents by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(y="City", data=df, order=df["City"].value_counts().iloc[:10].index)
plt.title("Top 10 Cities for Accidents")
plt.xlabel("Number of Accidents")
plt.ylabel("City")
plt.show()
```
